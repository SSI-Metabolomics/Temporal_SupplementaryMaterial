## Univariate correlation at the individual metabolite level using Spearman's rank correlation coefficient

To identify metabolic features correlating significantly with year of sampling we performed a Spearman's rank correlation coefficient test.

load libraries

```r
library(beeswarm)
library(ggplot2)
sessionInfo()
```

load blank filtered and imputed feature table

```r
ft <- read.csv('output/RAPCA_ImputedData_Rank100_NA_5.tsv', sep = '\t', check.names = F)
```

```r
dim(ft)
```

load metadata

```r
md <- read.table('output/MetaData_timsTOF_Qiime.tsv', sep = '\t', header = T, check.names = F, comment.char = "")
```

```r
dim(md)
```

are there any features with all zeros?

```r
length(which(rowSums(ft[,2:ncol(ft)]) == 0))
```

transpose feature table for correlation analysis

```r
ft <- as.data.frame(t(ft))
colnames(ft) <- ft[1,]
ft <- ft[-1,]
```

match metadata with feature table and bring filenames in same order (this should return TRUE)

```r
md <- md[which(md$'#SampleID' %in% rownames(ft)),]
md <- md[match(rownames(ft),md$'#SampleID'),]
identical(as.character(md$'#SampleID'),as.character(rownames(ft)))
```

select metadata category for correlation analysis

```r
catscont <- md$year
```

how many samples are there per year?

```r
table(catscont)
```

```r
ftt <- ft
```

assess correlation for one metabolite

```r
cor.test(catscont, scale(ftt[,2])[,1], method = 'spearman')
```

assess correlation for all metabolites and adjust p-values for multiple hypothesis testing

```r
corres <- t(sapply(1:ncol(ftt), function(x) unlist(cor.test(catscont, scale(ftt[,x])[,1], method = 'spearman')[c("estimate", "p.value")])))
corres <- cbind(corres,p.adjust(corres[,2], method = "fdr"))
```

```r
rownames(corres) <- colnames(ftt)
corres <- cbind(corres,rownames(corres))
colnames(corres) <- c('estimate.cor','p.value','p.value.corrected','cluster id')
```

```r
head(corres)
```

save correlation coefficients and p-values to file for import into Cytoscape

```r
write.table(corres, 'output/SpearmanCorrelation_year.txt', sep = '\t', quote = F, row.names = F)
```

how many features are significantly correlated with gestational age after FDR correction?

```r
corressig <- corres[which(as.numeric(corres[,3]) < 0.05),]
corressig <- corressig[order(corressig[,1]),]
```

```r
dim(corressig)
nrow(corressig)/ncol(ft)
```
